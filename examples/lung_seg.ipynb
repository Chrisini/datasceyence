## Preprocessing functions

```python
def get_lung(idx, src, side = "up", slice_n=25, resize=None, path="", mod=20):
    # resize = (100, 100)
    
    # ********** # begin plotting
    if idx % mod == 0:
        f, axarr = plt.subplots(1,7, figsize=(20,30))
    
    
    # ********** # manipulate src image
    # resize
    if resize is not None:
        src = cv2.resize(src, resize, interpolation = cv2.INTER_AREA)
    a = src.shape[0]
    b = a/2
    
    # manipulation
    src = cv2.normalize(src, None, alpha = 0, beta = 255, norm_type = cv2.NORM_MINMAX, dtype = cv2.CV_32F)
    src = src.astype(np.uint8)
    src = cv2.cvtColor(src, cv2.COLOR_GRAY2BGR)
    
    if idx % mod == 0:
        i = 0
        axarr[i].imshow(src)
        axarr[i].set_title("src")

    # ********** # thresholded image (ignoring where circle mask is zero)
    
    gray_image = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
    circle_mask = cv2.inRange(gray_image, 100, 255)
    
    if idx % mod == 0:
        i+=1
        axarr[i].imshow(circle_mask, cmap=plt.cm.gray)
        axarr[i].set_title("circle mask")
    
    ret, _ = cv2.threshold(gray_image[circle_mask!=0], 150, 255, cv2.THRESH_OTSU)
    print(f"threshold value for {side}: {ret}")
    thresh = gray_image < ret
    thresh = thresh.astype(np.uint8)
    
    if idx % mod == 0:
        i+= 1
        axarr[i].imshow(thresh, cmap=plt.cm.gray)
        axarr[i].set_title("threshold")
    
    # ********** # flood fill image edges black
    
    cv2.floodFill(thresh, None, (0, 0), (0))
    cv2.floodFill(thresh, None, (a-1, 0), (0))
    cv2.floodFill(thresh, None, (0, a-1), (0))
    cv2.floodFill(thresh, None, (a-1, a-1), (0))
    
    if idx % mod == 0:
        i+= 1
        axarr[i].imshow(thresh, cmap=plt.cm.gray)
        axarr[i].set_title("fill edges")
    
    # ********** # mask for up down left right
    
    side_mask = np.zeros(gray_image.shape, gray_image.dtype)
    if side == "up":
        poly = np.array([ [0,0], [0,b], [a,b], [a,0]], np.int32)
    elif side == "down":
        poly = np.array([ [0,b], [0,a], [a,a], [a,b]], np.int32)
    elif side == "left":
        poly = np.array([ [0,0], [b,0], [b,a], [0,a]], np.int32)
    elif side == "right":
        poly = np.array([ [b,0], [a,0], [a,a], [b,a]], np.int32)
        
    cv2.fillPoly(side_mask, [poly], (255))
    
    if idx % mod == 0:
        i+=1
        axarr[i].imshow(side_mask, cmap=plt.cm.gray)
        axarr[i].set_title("side mask")
    
    # ********** # bitwise end of thresholded image and mask
    
    thresh = cv2.bitwise_and(thresh, thresh, mask = side_mask)
     
    if idx % mod == 0:
        i+=1
        axarr[i].imshow(thresh, cmap=plt.cm.gray)
        axarr[i].set_title("combined")
    
    # ********** # get biggest connected component
    
    num_labels, labels_im = cv2.findContours(thresh,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)    
    
    if len(num_labels) > 0:
        max_area = max(num_labels, key = cv2.contourArea)
        cv2.drawContours(thresh,[max_area],0,255,-1)
    #gray = cv2.bitwise_not(thresh)
    
    mask = thresh.copy()
    
    #gray = cv2.inverse(gray)
    
    if idx % mod == 0:
        i+=1
        axarr[i].imshow(mask, cmap=plt.cm.gray)
        axarr[i].set_title("connected comp")
    
    # ********** # remove axis
    
    if idx % mod == 0:
        for a in range(0,i+1):
            axarr[a].axis('off')
        plt.show()

    return mask
    

img = nib.load(path)
src = nib.load(path).get_fdata()

mod = 20

for i in range(img.shape[2]):
    
    print("slice", i)
    
    # load image
    slice_i = src[:,:,i].copy()
    
    # get two parts of the lung
    up = get_lung(i, slice_i, side="up", slice_n=i, resize=resize, path=path, mod=mod)
    down = get_lung(i, slice_i, side="down", slice_n=i, resize=resize, path=path, mod=mod)

    final_mask = cv2.bitwise_or(up, down)
    
    # cv2.bitwise_and(slice_i, slice_i, mask=final_mask)
    roi = slice_i * final_mask
    
    src[:,:,i] = roi
    
    if i % mod == 0:
        plt.figure()
        plt.imshow(slice_i)
        plt.show()

        plt.figure()
        plt.imshow(final_mask)
        plt.show()

        plt.figure()
        plt.imshow(roi)
        plt.show()

if True:
    save_path = path.replace("data", "segmented")

    niii = nib.Nifti1Image(src, img.affine, img.header)
    nib.save(niii, save_path)


print("Done")

```

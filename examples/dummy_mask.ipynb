```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
height = 28
width = 6
max_diff = 3

# Initialize the array
array = np.zeros((height, width), dtype=int)

# Generate each line with noise, starting with a random position for the first pixel
lines = []
for i in range(width):
    line = [np.random.randint(0, height)]
    for j in range(1, height):
        # Generate the next point within the max difference range
        next_point = line[-1] + np.random.randint(-max_diff, max_diff + 1)
        next_point = np.clip(next_point, 0, height - 1)  # Ensure it stays within 0-27
        line.append(next_point)
    lines.append(line)

# Convert the list of lines to a numpy array of shape (28, 6)
array = np.array(lines).T  # Transpose to match the desired shape (28, 6)

# Plot each line on a 28x28 canvas
plt.figure(figsize=(5, 5))
for i in range(width):
    plt.plot(range(height), array[:, i], label=f"Line {i+1}")

plt.ylim(0, height - 1)
plt.xlabel("Pixel Position")
plt.ylabel("Line Value (0-27)")
plt.title("Noisy Horizontal Lines")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
height = 28
width = 6
max_diff = 2  # Smaller noise for more straightness
initial_spacing = 4  # Minimum vertical spacing between lines

# Initialize the array
array = np.zeros((height, width), dtype=int)

# Generate each line with noise
lines = []
start_y = 2  # Initial Y position for the first line

for i in range(width):
    line = [start_y]
    for j in range(1, height):
        # Generate the next point within the max difference range
        next_point = line[-1] + np.random.randint(-max_diff, max_diff + 1)
        next_point = np.clip(next_point, 0, height - 1)  # Ensure it stays within 0-27
        line.append(next_point)
    
    # Ensure no overlap by keeping a minimum spacing
    lines.append(line)
    start_y += initial_spacing  # Increase starting point for next line

# Convert the list of lines to a numpy array of shape (28, 6)
array = np.array(lines).T  # Transpose to match the desired shape (28, 6)

# Plot each line on a 28x28 canvas
plt.figure(figsize=(5, 5))
for i in range(width):
    plt.plot(range(height), array[:, i], label=f"Line {i+1}")

plt.ylim(0, height - 1)
plt.xlabel("Pixel Position")
plt.ylabel("Line Value (0-27)")
plt.title("Non-overlapping Slightly Straight Horizontal Lines")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MaxNLocator

# Fixing random state for reproducibility
np.random.seed(19)

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(projection='3d')

n = 100

colour = ['#BBBBBB', '#555555', '#BBBBBB', '#555555']

# For each set of style and range settings, plot n random points in the box
# defined by x in [23, 32], y in [0, 100], z in [zlow, zhigh].

m_max = 3
n_max = 3
l_max = 4

for l in range (l_max):
    
    x = np.array([[0, 0], [m_max-1, m_max-1]])
    y = np.array([[0, n_max-1], [0, n_max-1]])
    z = np.array([[l, l], [l, l]])
    
    ax.plot_surface(x, y, z, color=colour[l], alpha=0.3)

    if l==3: marker = 's'
    if l==2: marker = '^'
    if l==1: marker = 'o'
    if l==0: marker = '*'
    for m in range(3):
        for n in range(3):
            ax.scatter(m, n, l, marker=marker, c=colour[l], s=100)
            

ax.set_xlabel('M [index m]')
ax.set_ylabel('N [index n]')
ax.set_zlabel('L [index l]')

ax.scatter(0, 0, 2, label='F(m,n,l+1)', marker='^', c='tab:orange', s=120)
ax.scatter(0, 0, 1, label='F(m,n,l)', marker='o', c='tab:cyan', s=120)
ax.scatter(0, 0, 0, label='F(m,n,l-1)', marker='*', c='tab:pink', s=150)
ax.legend()

ax.set_zticklabels(['', 'L1', 'L2', 'L3', '1x1'])


ax.xaxis.set_major_locator(MaxNLocator(integer=True))
ax.yaxis.set_major_locator(MaxNLocator(integer=True))
ax.zaxis.set_major_locator(MaxNLocator(integer=True))

#ax.set_xticklabels(range(0, m_max))
#ax.set_yticklabels(range(0, n_max))
#ax.set_zticklabels(range(0, l_max))

ax.view_init(10, 220)

# ax.view_init(15, 240)

# ax.view_init(200, +80)

# ax.view_init(15, -15)
plt.tight_layout()
plt.savefig("example_results/layers.png")
plt.show()
```

```python
# Import libraries
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Create axis
axes = [3, 3, 5]
 
# Create Data
data = np.ones(axes)
 
# Control Transparency
alpha = 0.9
 
# Control colour
colors = np.ones(axes + [4])
 
colors[:,:,0] = [1, 0, 0, alpha]  # red
colors[:,:,1] = [0, 1, 0, alpha]  # green
colors[:,:,2] = [0, 0, 1, alpha]  # blue
colors[:,:,3] = [1, 1, 0, alpha]  # yellow
colors[:,:,4] = [1, 1, 1, alpha]  # grey
 
# Plot figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
# Voxels is used to customizations of
# the sizes, positions and colors.
ax.voxels(data, facecolors=colors, edgecolors='grey')

ax.xaxis.set_major_locator(MaxNLocator(integer=True))
ax.yaxis.set_major_locator(MaxNLocator(integer=True))
ax.zaxis.set_major_locator(MaxNLocator(integer=True))
# ax.view_init(120, -290)

ax.set_title('F(m,n,l)')
ax.set_zlabel('C')

print("done")
```

```python
import matplotlib.pyplot as plt
import numpy as np


# prepare some coordinates
x, y, z = np.indices((8, 8, 8))

# draw cuboids in the top left and bottom right corners, and a link between
# them
cube1 = (x < 3) & (y < 3) & (z < 3)
cube2 = (x >= 5) & (y >= 5) & (z >= 5)
link = abs(x - y) + abs(y - z) + abs(z - x) <= 2

# combine the objects into a single boolean array
voxelarray = cube1 | cube2 | link

# set the colors of each object
colors = np.empty(voxelarray.shape, dtype=object)
colors[link] = 'red'
colors[cube1] = 'blue'
colors[cube2] = 'green'

# and plot everything
ax = plt.figure().add_subplot(projection='3d')
ax.voxels(voxelarray, facecolors=colors, edgecolor='k')

plt.show()
```
